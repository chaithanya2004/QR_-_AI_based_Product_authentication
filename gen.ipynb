```python
from GoogleImageScraper import GoogleImageScraper

webdriver_path = "chromedriver.exe"
image_path = "dataset/Nivea"
search_keys = ["Nivea cream original", "Nivea product front", "Nivea barcode", "Nivea packaging"]
num_images = 100

scraper = GoogleImageScraper(webdriver_path, image_path, search_keys, num_images)
scraper.find_image_urls()
scraper.save_images()
```
