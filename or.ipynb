```python
import os
import shutil

SOURCE_BENIGN = "non_qr/benign"
SOURCE_MALICIOUS = "non_qr/malicious"
DEST = "non_qr"

# Create destination folder if not exists
os.makedirs(DEST, exist_ok=True)

def copy_from_folder(folder):
    if not os.path.exists(folder):
        print(f"Folder not found: {folder}")
        return

    count = 0
    for file in os.listdir(folder):
        if file.lower().endswith((".jpg", ".jpeg", ".png")):
            shutil.copy(os.path.join(folder, file), DEST)
            count += 1

    print(f"Copied {count} images from {folder}")

copy_from_folder(SOURCE_BENIGN)
copy_from_folder(SOURCE_MALICIOUS)

print(f"All images moved to {DEST}. Dataset ready for training!")
```
